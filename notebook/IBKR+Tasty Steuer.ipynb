```python
import pandas as pd
import csv
from currency_converter import ECB_URL, CurrencyConverter
from datetime import date
```

# IBKR

```python
with open("MULTI_20230102_20230602.csv", newline="") as csvfile:
    data = csv.reader(csvfile, delimiter=",", quotechar="\"")
    transactions = []
    for row in data:
        if row[0] == "Transaktionen":
            transactions.append(row)
```

```python
df = pd.DataFrame.from_records(transactions[1:], columns=transactions[0])
df
```

Data cleaning

```python
df_clean = df[(df.Header == "Data") & (df.Vermögenswertkategorie == "Aktien- und Indexoptionen")]
```

```python
df_clean["Datum/Zeit"] = pd.to_datetime(df_clean["Datum/Zeit"], format="%Y-%m-%d, %H:%M:%S")
```

```python
df_clean
```

Categorize transactions

```python
df_clean["Menge"] = df_clean["Menge"].astype(float)
df_clean["Erträge"] = df_clean["Erträge"].astype(float)
df_clean["Prov./Gebühr"] = df_clean["Prov./Gebühr"].astype(float)
```

```python
df_clean.loc[(df_clean["Menge"] > 0) & (df_clean["Code"].str.contains("O")), "Typ"] = "BTO"   # open long
df_clean.loc[(df_clean["Menge"] < 0) & (df_clean["Code"].str.contains("C")), "Typ"] = "STC"   # close long

df_clean.loc[(df_clean["Menge"] < 0) & (df_clean["Code"].str.contains("O")), "Typ"] = "STO"   # open short
df_clean.loc[(df_clean["Menge"] > 0) & (df_clean["Code"].str.contains("C")), "Typ"] = "BTC"   # close short
```

```python
df_clean
```

```python
df_clean.loc[df_clean["Typ"] == "STC", "GuV Termingeschäft"] = df_clean["Realisierter G&V"]
df_clean.loc[(df_clean["Typ"] == "STO") | (df_clean["Typ"] == "BTC"), "GuV Stillhaltergeschäft"] = df_clean["Erträge"] + df_clean["Prov./Gebühr"]
```

```python
df_clean
```

Currency conversion

```python
# https://www.ecb.europa.eu/stats/eurofxref/eurofxref-hist.zip
```

```python
# test
c = CurrencyConverter(ECB_URL, fallback_on_missing_rate=True)
c.convert(100, 'EUR', 'USD', date=date(2023, 5, 2))
```

```python
df_clean["GuV Termingeschäft EUR"] = df_clean.apply(
    lambda x: c.convert(
        x["GuV Termingeschäft"], 
        'EUR', 'USD', 
        date=x["Datum/Zeit"].date()), 
    axis=1)

df_clean["GuV Stillhaltergeschäft EUR"] = df_clean.apply(
    lambda x: c.convert(
        x["GuV Stillhaltergeschäft"], 
        'EUR', 'USD', 
        date=x["Datum/Zeit"].date()), 
    axis=1)
```

```python
df_clean
```

### KAP

Zeile 21: Stillhaltereinkünfte und Gewinne aus Termingeschäften

```python
df_clean["GuV Stillhaltergeschäft EUR"].sum() + df_clean.loc[df_clean["GuV Termingeschäft EUR"] > 0, "GuV Termingeschäft EUR"].sum()
```

Zeile 24: Verluste aus Termingeschäften

```python
df_clean.loc[df_clean["GuV Termingeschäft EUR"] < 0, "GuV Termingeschäft EUR"].sum()
```

# Tasty

```python
df_tasty = pd.read_csv("2023-5WX64887-gain-loss.csv")
df_tasty
```

```python
for col in ["NO_WS_COST", "NO_WS_PROCEEDS", "NO_WS_GAINLOSS"]:
    df_tasty[col] = df_tasty[col].str.replace("$", "").astype(float)
```

```python
for col in ["OPEN_DATE", "CLOSE_DATE"]:
    df_tasty[col] = pd.to_datetime(df_tasty[col], format="%Y-%m-%d")
```

```python
df_tasty.loc[df_tasty["LONG_SHORT_IND"] == "L", "GuV Termingeschäft"] = df_tasty["NO_WS_GAINLOSS"]
df_tasty.loc[df_tasty["LONG_SHORT_IND"] == "S", "GuV Stillhaltergeschäft open"] = df_tasty["NO_WS_PROCEEDS"]
df_tasty.loc[df_tasty["LONG_SHORT_IND"] == "S", "GuV Stillhaltergeschäft close"] = -df_tasty["NO_WS_COST"]
```

```python
df_tasty["GuV Termingeschäft EUR"] = df_tasty.apply(
    lambda x: c.convert(
        x["GuV Termingeschäft"], 
        'EUR', 'USD', 
        date=x["CLOSE_DATE"].date()), 
    axis=1)

df_tasty["GuV Stillhaltergeschäft open EUR"] = df_tasty.apply(
    lambda x: c.convert(
        x["GuV Stillhaltergeschäft open"], 
        'EUR', 'USD', 
        date=x["OPEN_DATE"].date()), 
    axis=1)

df_tasty["GuV Stillhaltergeschäft close EUR"] = df_tasty.apply(
    lambda x: c.convert(
        x["GuV Stillhaltergeschäft close"], 
        'EUR', 'USD', 
        date=x["CLOSE_DATE"].date()), 
    axis=1)
```

```python
df_tasty
```

### KAP

Zeile 21: Stillhaltereinkünfte und Gewinne aus Termingeschäften

```python
df_tasty["GuV Stillhaltergeschäft open EUR"].sum() + \
df_tasty["GuV Stillhaltergeschäft close EUR"].sum() + \
df_tasty.loc[df_tasty["GuV Termingeschäft EUR"] > 0, "GuV Termingeschäft EUR"].sum()
```

Zeile 24: Verluste aus Termingeschäften

```python
df_tasty.loc[df_tasty["GuV Termingeschäft EUR"] < 0, "GuV Termingeschäft EUR"].sum()
```
